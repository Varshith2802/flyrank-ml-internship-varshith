## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
# Check why a ranked queue is needed

REVIEW_CAPACITY = 50
PRIMARY_METRIC = "Precision@50"

review_share = REVIEW_CAPACITY / len(df)

print("Pages available:", f"{len(df):,}")
print("Initial review capacity:", REVIEW_CAPACITY)
print("Share reviewed in the first queue:", f"{review_share:.3%}")
print("Provisional evaluation metric:", PRIMARY_METRIC)
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
# Calculate real supporting numbers from the starter dataset

declining_mask = (
    df["trend_direction"]
    .astype(str)
    .str.lower()
    .eq("down")
)

pages_with_impressions = df[df["impressions_90d"] > 0].copy()

correlation = pages_with_impressions["search_volume"].corr(
    pages_with_impressions["impressions_90d"]
)

print("SUPPORTING NUMBERS")
print("-" * 40)
print("Total pages:", f"{len(df):,}")
print(
    "Declining pages:",
    f"{declining_mask.sum():,}",
    f"({declining_mask.mean():.1%})"
)
print(
    "Pages with impressions above zero:",
    f"{len(pages_with_impressions):,}"
)
print(
    "Search volume/impressions correlation:",
    f"{correlation:.3f}"
)
```

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

```python
# Document the main leakage and claim-safety rules

TARGET_PROXY = "trend_direction == 'down'"

FORBIDDEN_PREDICTOR_COLUMNS = [
    "trend_direction",
    "trend_pct"
]

IDENTIFIER_COLUMNS = [
    "content_id",
    "client_id"
]

print("Current starter proxy:", TARGET_PROXY)

print("\nColumns excluded because they reveal the target:")
for column in FORBIDDEN_PREDICTOR_COLUMNS:
    print("-", column)

print("\nIdentifiers used only for grouping or validation:")
for column in IDENTIFIER_COLUMNS:
    print("-", column)

required_columns = (
    FORBIDDEN_PREDICTOR_COLUMNS + IDENTIFIER_COLUMNS
)

missing = [
    column for column in required_columns
    if column not in df.columns
]

print("\nSafety check:")
if missing:
    print("Missing expected columns:", missing)
else:
    print("All expected safety-check columns are present.")
    print("They will not be used as ordinary model features.")
```
